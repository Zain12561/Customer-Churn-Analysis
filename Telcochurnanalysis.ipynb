```python
def conv(value):
    if value == 1:
        return "Yes"
    else:
        return "No"
df['SeniorCitizen']=df['SeniorCitizen'].apply(conv)
```

```python
plt.figure(figsize = (3,4))
gb = df.groupby('Churn').agg({'Churn':"count"})
plt.pie(gb['Churn'], labels = gb.index, autopct = "%1.2f%%")
plt.title('Percentage of Customers ', fontsize= 10)
plt.show()
```

```python
plt.figure(figsize =(3,4))
sns.countplot(x = "SeniorCitizen" , data = df , hue = 'Churn')
plt.title('Churn by senior citizen')
plt.show()
```

```python
grouped = df.groupby(['SeniorCitizen', 'Churn']).size().unstack(fill_value=0)

percentages = grouped.div(grouped.sum(axis=1), axis=0) * 100

# Step 3: Plot
plt.figure(figsize=(3, 4))
bottom = [0, 0]  # For stacking (assuming two groups: 0 and 1 in SeniorCitizen)

for churn_status in percentages.columns:
    values = percentages[churn_status].values
    plt.bar([0, 1], values, bottom=bottom, label=churn_status)

    # Add percentage labels
    for i in range(len(values)):
        plt.text(i, bottom[i] + values[i]/2, f'{values[i]:.1f}%', ha='center', va='center', fontsize=9)

    bottom = [bottom[i] + values[i] for i in range(len(values))]

# X-axis labels
plt.xticks([0, 1], ['No', 'Yes'])  # 0 = Not Senior, 1 = Senior
plt.title('Churn by senior citizen')
plt.ylabel('Percentage')
plt.legend(title='Churn')
plt.tight_layout()
plt.show()
```

```python
# Assuming your dataframe is called df
columns = ['PhoneService', 'MultipleLines', 'InternetService',
           'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
           'TechSupport', 'StreamingTV', 'StreamingMovies']

# Set up the matplotlib figure
n_cols = 3
n_rows = (len(columns) + n_cols - 1) // n_cols  # Calculate needed rows
fig, axes = plt.subplots(n_rows, n_cols, figsize=(9, 15))
axes = axes.flatten()  # Flatten in case it's 2D

for i, col in enumerate(columns):
    sns.countplot(x=col, data=df, ax=axes[i]  , hue =df['Churn'])
    axes[i].set_title(f'{col} Count')
    axes[i].tick_params(axis='x', rotation=45)  # Rotate x-labels if needed

# Hide any unused subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

#it is seen that people who are using Stramng TV Count and Streaming Movie service are usually more churn out.People using tech Support and online securirt are minimum to churn out . We are only taking about those customers who using services.
